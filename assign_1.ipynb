```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#APPLE

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
df = pd.read_csv('AAPL.csv',parse_dates=True)
df["Index"] = df.index
df['Volume']=df['Volume'].astype(float)
x = df[['Index','Close']]
y = df['Adj Close']
train_x, test_x, train_y, test_y = train_test_split(x,y,test_size = 0.2,random_state = 10,shuffle=False)
scal = StandardScaler().fit(train_x)


lr = LinearRegression()
lr.fit(train_x,train_y)
pred1 = lr.predict(test_x)
fig = plt.figure(figsize =(20,10))
plt.plot(test_x.index,test_y)
plt.plot(test_x.index,pred1)
plt.legend(["Adj Close","Predicted Adj Close"],loc ="lower left")
plt.show()


dr = DecisionTreeRegressor(random_state=0)
dr.fit(train_x,train_y)
pred4 = dr.predict(test_x)
fig = plt.figure(figsize =(20,10))
plt.plot(test_x.index,test_y)
plt.plot(test_x.index,pred4)
plt.legend(["Adj Close","Predicted Adj Close"],loc ="lower left")
plt.show()


els = ElasticNet(random_state=0)
els.fit(train_x,train_y)
pred3 = els.predict(test_x)
fig = plt.figure(figsize =(20,10))
plt.plot(test_x.index,test_y)
plt.plot(test_x.index,pred3)
plt.legend(["Adj Close","Predicted Adj Close"],loc ="lower left")
plt.show()


kn = KNeighborsRegressor(n_neighbors=2)
kn.fit(train_x,train_y)
pred5 = kn.predict(test_x)
fig = plt.figure(figsize =(20,10))
plt.plot(test_x.index,test_y)
plt.plot(test_x.index,pred5)
plt.legend(["Adj Close","Predicted Adj Close"],loc ="lower left")
plt.show()


las = linear_model.Lasso(alpha=0.1)
las.fit(train_x,train_y)
pred6 = las.predict(test_x)
fig = plt.figure(figsize =(20,10))
plt.plot(test_x.index,test_y)
plt.plot(test_x.index,pred6)
plt.legend(["Adj Close","Predicted Adj Close"],loc ="lower left")
plt.show()

```
